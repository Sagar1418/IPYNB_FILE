```python
import os
import pandas as pd
from collections import Counter

# Set your data folder path
folder_path = '200/200'

# Initialize a Counter to collect voltage values
voltage_counter = Counter()

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        try:
            df = pd.read_csv(file_path)

            # Case-insensitive voltage column search
            voltage_col = [col for col in df.columns if col.lower() == 'voltage']
            if voltage_col:
                voltage_values = df[voltage_col[0]].dropna()
                voltage_counter.update(voltage_values)
        except Exception as e:
            print(f"Could not read {filename}: {e}")

# Print unique voltage values and their counts
print("Unique voltage values and their counts across all files:")
for voltage, count in voltage_counter.items():
    print(f"{voltage}: {count}")
```

```python
import pandas as pd
import os

# Folder where all your data files are stored
data_folders = ['200/200' , '200-400/200-400']

# Column to extract from each file
column_name = 'SWNO'

# List to collect switch_ids from all files
all_switch_ids = []

# Loop over all files in the folder
for data_folder in data_folders:
    for filename in os.listdir(data_folder):
        if filename.endswith('.csv'):
            file_path = os.path.join(data_folder, filename)
            try:
                df = pd.read_csv(file_path)
                if column_name in df.columns:
                    all_switch_ids.extend(df[column_name].dropna().tolist())
            except Exception as e:
                print(f"Error reading {filename}: {e}")

```

```python
# Get unique switch IDs
unique_switch_ids = set(all_switch_ids)

# Load the reference CSV file with SOURCE_SWITCH_ID column
source_file = 'HTCABLE.csv'
source_df = pd.read_csv(source_file, dtype={'SOURCE_SWITCH_ID': str}, low_memory=False)
source_df['SOURCE_SWITCH_ID'] = source_df['SOURCE_SWITCH_ID'].astype(str).str.strip()

# Match and unmatched
matched_df = source_df[source_df['SOURCE_SWITCH_ID'].isin(unique_switch_ids)]
unmatched_df = source_df[~source_df['SOURCE_SWITCH_ID'].isin(unique_switch_ids)]

# Print results
print(f"✅ Total SOURCE_SWITCH_IDs in reference file: {source_df['SOURCE_SWITCH_ID'].nunique()}")
print(f"🔗 Number of unique matches: {matched_df['SOURCE_SWITCH_ID'].nunique()}")
print(f"❌ Number of unique non-matches: {unmatched_df['SOURCE_SWITCH_ID'].nunique()}")
```

```python
import polars as pl
from pathlib import Path

data_folders = [
    "/media/sagarkumar/New Volume/SAGAR/200/200",
    "/media/sagarkumar/New Volume/SAGAR/200-400/200-400",
    # "/media/sagarkumar/New Volume/SAGAR/400-600/400-600",
    # "/media/sagarkumar/New Volume/SAGAR/600-759/600-759",
]

all_csvs = []
for folder in data_folders:
    fp = Path(folder)
    if fp.is_dir():
        all_csvs.extend(sorted(fp.glob("*.csv")))

if not all_csvs:
    raise SystemExit("No CSV files found.")

# --- Force all columns to be "str" (fixes schema mismatch) ---
def force_string_schema(path):
    # Read just the header for column names
    import csv
    with open(path, newline='') as f:
        reader = csv.reader(f)
        header = next(reader)
    # Build string schema
    schema = {col: pl.String for col in header}
    return schema

scans = [
    pl.scan_csv(
        csv,
        dtypes=force_string_schema(csv),
        infer_schema_length=0   # don't guess, just take as string
    ).with_columns(
        pl.lit(csv.name).alias("__source_file")
    )
    for csv in all_csvs
]

df = pl.concat(scans, how="diagonal").collect(streaming=True)
df.write_parquet("/media/sagarkumar/New Volume/SAGAR/combined_scada.parquet")
print(" Parquet saved to /media/sagarkumar/New Volume/SAGAR/combined_scada.parquet")
df.write_csv("/media/sagarkumar/New Volume/SAGAR/combined_scada.csv")
print(" CSV saved to /media/sagarkumar/New Volume/SAGAR/combined_scada.csv")
```

```python
import pandas as pd
from pathlib import Path
import multiprocessing as mp

# Paths as before
data_folders = [
    "/media/sagarkumar/New Volume/SAGAR/200/200",
    "/media/sagarkumar/New Volume/SAGAR/200-400/200-400",
    "/media/sagarkumar/New Volume/SAGAR/400-600/400-600",
    "/media/sagarkumar/New Volume/SAGAR/600-759/600-759",
]

# Gather all CSVs
csv_files = []
for folder in data_folders:
    fp = Path(folder)
    if fp.is_dir():
        csv_files.extend(fp.glob("*.csv"))

def get_unique_swno(file):
    try:
        # Try utf-8 first, fallback to latin1
        try:
            col = pd.read_csv(file, usecols=["SWNO"], dtype=str, encoding='utf-8', low_memory=False).SWNO
            print(f"Processed {file} successfully with utf-8 encoding.")
        except UnicodeDecodeError:
            col = pd.read_csv(file, usecols=["SWNO"], dtype=str, encoding='latin1', low_memory=False).SWNO
            print(f"Processed {file} successfully with latin1 encoding.")
        # Return unique values as set
        return set(col.dropna().unique())
    except Exception as e:
        print(f"Failed to process {file}: {e}")
        return set()

# Multiprocessing pool
with mp.Pool(processes=mp.cpu_count()) as pool:
    results = pool.map(get_unique_swno, csv_files)

# Union of all sets
all_unique_swno = set().union(*results)

print(f"Found {len(all_unique_swno):,} unique SWNO values across all files.")
# Save to CSV if desired:
outpath = "/media/sagarkumar/New Volume/SAGAR/all_unique_SWNO.csv"
pd.Series(sorted(all_unique_swno)).to_csv(outpath, index=False, header=["SWNO"])
print(f"Saved unique SWNO values to {outpath}")
```

```python
import pandas as pd
from pathlib import Path
import multiprocessing as mp

data_folders = [
    "/media/sagarkumar/New Volume/SAGAR/200/200",
    "/media/sagarkumar/New Volume/SAGAR/200-400/200-400",
    "/media/sagarkumar/New Volume/SAGAR/400-600/400-600",
    "/media/sagarkumar/New Volume/SAGAR/600-759/600-759",
]

# collect every CSV path once
csv_files = [p for folder in data_folders
               for p in Path(folder).glob("*.csv")
               if Path(folder).is_dir()]

def unique_swno_from_one(csv_path: Path):
    """Return filename + its unique SWNO set (empty set if missing/error)."""
    try:
        try:
            col = pd.read_csv(csv_path,
                              usecols=["SWNO"],
                              dtype=str,
                              encoding="utf-8",
                              low_memory=False).SWNO
            print(f"  {csv_path.name} processed with utf-8 encoding")
        except UnicodeDecodeError:
            col = pd.read_csv(csv_path,
                              usecols=["SWNO"],
                              dtype=str,
                              encoding="latin1",
                              low_memory=False).SWNO
            print(f"  {csv_path.name} processed with latin1 encoding")
        uniques = set(col.dropna().unique())
        return csv_path.name, uniques
    except Exception as e:
        print(f"  {csv_path.name} skipped ({e})")
        return csv_path.name, set()

# parallel scan
with mp.Pool(mp.cpu_count()) as pool:
    per_file_results = pool.map(unique_swno_from_one, csv_files)

# ───────────────────────────────────────────────
# 1) master union for all files
# ───────────────────────────────────────────────
all_unique = set().union(*(s for _, s in per_file_results))
pd.Series(sorted(all_unique), name="SWNO").to_csv(
    "/media/sagarkumar/New Volume/SAGAR/all_unique_SWNO.csv",
    index=False
)
print(f"master list: {len(all_unique):,} unique SWNO saved to all_unique_SWNO.csv")

# ───────────────────────────────────────────────
# 2) per-file unique list
# ───────────────────────────────────────────────
rows = []
for fname, sw_set in per_file_results:
    if sw_set:                                # ignore empty/missing cases
        rows.append({"source_file": fname,
                     "unique_SWNO": ";".join(sorted(sw_set))})
per_file_df = pd.DataFrame(rows)

per_file_df.to_csv(
    "/media/sagarkumar/New Volume/SAGAR/each_file_unique_SWNO.csv",
    index=False
)
print(f" per-file list saved to each_file_unique_SWNO.csv "
      f"({len(per_file_df):,} files written)")
```
